```python
from sklearn import (
    neural_network,
)
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
df_treino = pd.read_csv('dados_teste.csv')
df_treino.head()

X_treino = df_treino.drop(columns=['class'])
y_treino = df_treino['class']
```

```python
df_original = pd.read_csv('dados_finais_300.csv')
df_original.head()

X_final = df_original.drop(columns=['class'])
y_final = df_original['class']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_final)
X_teste_scaled = scaler.transform(X_treino)
```

```python
MLP = neural_network.MLPClassifier(
    hidden_layer_sizes=(18, 10),
    solver='adam',
    learning_rate_init=0.01,
    momentum=0.1,
    max_iter=2000,
    verbose=True,
    random_state=42
    )
MLP.fit(X_scaled, y_final)
```

```python
MLP.predict(X_teste_scaled)
```

```python
y_pred_train = MLP.predict(X_scaled)
print("Relatório treino:")
print(classification_report(y_final, y_pred_train))

# Avaliação no teste
y_pred_test = MLP.predict(X_teste_scaled)
print("Relatório teste:")
print(classification_report(y_treino, y_pred_test))
```

```python
print("Classes: ", MLP.classes_ )
print("Melhor loss: ", MLP.best_loss_)
print(MLP.coefs_)
print(MLP.intercepts_)
print(MLP.get_params)
```
